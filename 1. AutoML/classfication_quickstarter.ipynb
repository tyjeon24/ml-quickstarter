```python
# 1. Prepare data

from pycaret.datasets import get_data # pip install pycaret==3.0.1
juice = get_data('juice')
juice.head()

target = "Purchase" # Y label. You need to change this value.
```

```python
# 2. Find the best model, no need to modify!

from pycaret.classification import *
s = setup(data = juice,  target = target)
top3 = compare_models(n_select = 3)
tuned_top3 = [tune_model(i, choose_better=True) for i in top3]
blender = blend_models(tuned_top3)
stacker = stack_models(tuned_top3)
best_auc_model = automl(optimize = 'AUC')
```

```python
# 3. Save model
final_model = finalize_model(best_auc_model)
save_model(final_model, "final_classification_model")
```
